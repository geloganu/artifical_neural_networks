```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

# Optimizing Alpha

```python
sig1 = 'sigmoid_data1.out1'
sig01 = 'sigmoid_data01.out1'
sig001 = 'sigmoid_data001.out1'
sig100 = 'sigmoid_data100.out1'
tanh1 = 'tanh_data1.out1'
tanh01 = 'tanh_data01.out1'
tanh001 = 'tanh_data001.out1'

s1_epoch, s1_error = np.loadtxt(sig1,delimiter=',',skiprows=1,unpack=True)
s01_epoch, s01_error = np.loadtxt(sig01,delimiter=',',skiprows=1,unpack=True)
s001_epoch, s001_error = np.loadtxt(sig001,delimiter=',',skiprows=1,unpack=True)
s100_epoch, s100_error = np.loadtxt(sig100,delimiter=',',skiprows=1,unpack=True)

t1_epoch, t1_error = np.loadtxt(tanh1,delimiter=',',skiprows=1,unpack=True)
t01_epoch, t01_error = np.loadtxt(tanh01,delimiter=',',skiprows=1,unpack=True)
t001_epoch, t001_error = np.loadtxt(tanh001,delimiter=',',skiprows=1,unpack=True)


XOR001 = 'XOR_sigmoid001.out1'
XOR01 = 'XOR_sigmoid01.out1'
XOR1 = 'XOR_sigmoid1.out1'
XOR10 = 'XOR_sigmoid10.out1'
XOR100 = 'XOR_sigmoid100.out1'

XOR001_epoch, XOR001_error = np.loadtxt(XOR001,delimiter=',',skiprows=1,unpack=True)
XOR01_epoch, XOR01_error = np.loadtxt(XOR01,delimiter=',',skiprows=1,unpack=True)
XOR1_epoch, XOR1_error = np.loadtxt(XOR1,delimiter=',',skiprows=1,unpack=True)
XOR10_epoch, XOR10_error = np.loadtxt(XOR10,delimiter=',',skiprows=1,unpack=True)
XOR100_epoch, XOR100_error = np.loadtxt(XOR100,delimiter=',',skiprows=1,unpack=True)


```

```python
fig = plt.figure(figsize=(5.5,5.5))

plt.plot(s1_epoch,s1_error, "b-",label='Sigmoid',linewidth=0.8)
plt.plot(t1_epoch,t1_error, "r-",label='Tanh',linewidth=0.8)
plt.plot(s01_epoch,s01_error, "b--",linewidth=0.8)
plt.plot(t01_epoch,t01_error, "r--",linewidth=0.8)
plt.plot(s001_epoch,s001_error, "b:",linewidth=0.8)
plt.plot(t001_epoch,t001_error, "r:",linewidth=0.8)
plt.plot(s100_epoch,s100_error, "r:",linewidth=0.8)

plt.yscale('log')
plt.xscale('log')

plt.ylabel("Error")
plt.xlabel("Training Epoch")
plt.title('Error vs. Epoch for Various Training Steps')

plt.legend()

#plt.savefig('training_step_epoch.png')
```

For sigmoid function, alpha=100 yielded the most optimal decrease in error. For tanh, it was for alpha=0.01

```python
tanha = np.array([0.001,0.01,0.1,1,10,100,1000])
tanht = np.array([8.288,6.137,6.154,5.280,4.831,4.393,4.800])
tanht *= 1E-03

siga = tanha
sigt = np.array([5.361,6.002,5.379,5.402,5.480,5.213,3.9498])
sigt *= 1E-02
```

```python
fig = plt.figure(figsize=(5.5,5.5))

plt.plot(siga,sigt, "b-",label='Sigmoid',linewidth=0.8)
plt.plot(tanha,tanht, "r-",label='Tanh',linewidth=0.8)


#plt.yscale('log')

plt.ylabel("Time")
plt.xlabel("Alpha")
plt.title('Batch Training Time for Different Training Step Value')

plt.legend()

plt.savefig('training_time.png')
```

```python
fig = plt.figure(figsize=(5.5,5.5))

plt.plot(XOR1_epoch,XOR1_error, "b",label='a=1',linewidth=0.8)
plt.plot(XOR01_epoch,XOR01_error, "r",label='a=0.1',linewidth=0.8)
plt.plot(XOR001_epoch,XOR001_error, "g",label='a=0.01',linewidth=0.8)
plt.plot(XOR100_epoch,XOR100_error, "m",label='a=100',linewidth=0.8)
plt.plot(XOR10_epoch,XOR10_error, "y",label='a=10',linewidth=0.8)



plt.yscale('log')
plt.xscale('log')

plt.ylabel("Error")
plt.xlabel("Training Epoch")
plt.title('Error vs. Epoch for Various Training Steps')

plt.legend()

plt.savefig('XOR_training_step_epoch.png')
```
